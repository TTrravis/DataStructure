# 树
## 二叉树的性质
1. 二叉树是每个节点最多有两个子树的树结构
2. 二叉树第k层上的节点数目**至多**有$2^{k-1}$
3. 深度为h的二叉树至多有$2^h-1$个节点
4. 包含n个节点的二叉树的高度至少为$log_2(n+1)$
5. 在任意一棵二叉树中，若叶子节点的个数为$n_0$，度为2的节点数为$n_2$,则$n_0=n_2+1$

## 常见的二叉树
1. 满二叉树
2. 完全二叉树
3. 平衡二叉树

```python
# python实现二叉树，每个节点都是node对象，通过tree类中的add()方法向二叉树中加入树节点，构成完全二叉树或非完全二叉树
# """
# 节点
# 在链表中，一个节点只能有一个后继节点和前驱节点；在树中，一个节点可以有多个后继节点，只能有一个前驱节点
# 节点的度表示有几个后继节点
# 在树中，没有前驱节点的节点称为根节点，没有后继节点的节点称为叶子节点
# """
class Node(object):
    def __int__(self,val=None,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right#每个节点有两个后继节点，分别为左子节点和右子节点
# """
# 构造树
# 每个树有一个根节点
# """
class Tree(object):
    def __int__(self,node=None):
        self.root=None#每个树有一个根节点
# """
# 层遍历
# 从根节点最高层到最低层，每层从左向右遍历。
# 需要借助队列实现
# """
    def level_travel(self):
        if self.root==None:
            return
        queue=[self.root]
        while queue:
            cur=node=queue.pop(0)
            print(cur.val, end=' ')
            if cur.left !=None:
                queue.append(cur.left)
            if cur.right !=None:
                queue.append(cur.right)
# """
# 添加节点
# 从第一层开始，依次判断根节点是否有左子节点和右子节点
# 没有则添加；有则转到第二层
# """
    def add(self,item=None):
        node=Node(val=item)
        if not self.root or self.root.val is None:
            self.root=node
        else:
            queue=[]
            queue.append(self.root)
            while True:
                current_node=queue.pop(0)
                if current_node.val is None:
                    continue
                if not current_node.left:
                    current_node.left=node
                    return
                elif not current_node.right:
                    current_node.right=node
                    return
                else:
                    queue.append(current_node.left)
                    queue.append(current_node.right)
    #先序遍历
    #先遍历根节点，再遍历左子树，最后遍历右子树
    def preorder_travel(self, node):
        #如果节点为空
        if node == Node:
            return
        print(node.val,end=' ')
        self.preorder_travel(node.left)
        self.preorder_travel(node.right)
        #递归
    #中序遍历
    #先遍历左子树，再遍历根节点，最后遍历右子树
    def inorder_travel(self,node):
        #如果节点为空
        if node == None:
            return
        self.inorder_travel(node.left)
        print(node.val,end=' ')
        self.inorder_travel(node.right)
    #后序遍历
    #先遍历左子树，再遍历右子树，最后遍历根节点
    def postorder_travel(self,node):
        # 如果节点为空
        if node==None:
            return
        self.postorder_travel(node.left)
        self.postorder_travel(node.right)
        print(node.val, end=' ')
```
